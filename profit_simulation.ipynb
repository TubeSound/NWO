```python
from random import random
import numpy as np
import pandas as pd
from bokeh.plotting import figure
from bokeh.plotting import show
from bokeh.plotting import output_notebook

output_notebook()

leverledge = 1000
red_line = -100    

def simulation(win_rate, profit_target, loss, lot):

    print(f'勝率: {win_rate * 100: .3f}%, 目標利益: ${profit_target}, ストップロス: ${loss}, ロット: {lot} ')
    print(f"証拠金:  ${loss * lot * leverledge} レバレッジ: {leverledge}") 
    
    profits = []
    s = 0
    for i in range(1000):
        r = random() 
        if r >= win_rate:
            profit = - loss * lot * leverledge
        else:
            profit = profit_target * lot * leverledge
        s += profit
        profits.append(s)
        
    p = figure(title="Profit curve", x_axis_label='n', y_axis_label='profit')
    p.line(range(len(profits)), profits, legend_label="Profit sum", line_width=2)
    p.line([0, len(profits) - 1], [red_line, red_line], color='red')
    show(p)
```

```python
simulation(0.2, 50, 10.0, 0.001)
```

```python
simulation(0.2, 50, 5.0, 0.002)
```

```python
simulation(0.3, 5, 2.0, 0.005)
```

```python
simulation(0.1, 100, 5.0, 0.008)
```
